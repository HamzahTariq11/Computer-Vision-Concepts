# Q7

```python
gray_image = cv2.imread('cv lab.jpg', cv2.IMREAD_GRAYSCALE)
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rotation_angle = 45

rotation_matrix = cv2.getRotationMatrix2D((binary_image.shape[1] / 2, binary_image.shape[0] / 2), rotation_angle, 1)

rotated_image = cv2.warpAffine(binary_image, rotation_matrix, (binary_image.shape[1], binary_image.shape[0]))

cv2_imshow(rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure(figsize=(10, 7))

rows = 2

col = 2

fig.add_subplot(rows, col,1)
plt.imshow(cv2.cvtColor(binary_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("First")

fig.add_subplot(rows, col,2)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Second")
```

##  Q8

```python
size=(300,300)
image1 = cv2.imread('batman.jpg')
image2 = cv2.imread('cv lab.jpg')
image1=cv2.resize(image1,size)
image2=cv2.resize(image2,size)
alpha = 0.5
blended_image = cv2.addWeighted(image1, alpha, image2, alpha, 0)
cv2_imshow(blended_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_image = cv2.cvtColor(blended_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
equalized_image = cv2.equalizeHist(gray_image)
cv2_imshow(equalized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure(figsize=(10, 7))

rows = 2

col = 3

fig.add_subplot(rows, col,1)
plt.imshow(cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("First")

fig.add_subplot(rows, col,2)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Second")

fig.add_subplot(rows, col,3)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Third")
```

## Q9

```python
image=cv2.imread('batman.jpg')
height,width,_=image.shape
cv2_imshow(image)
mask=np.zeros((height,width), dtype=np.uint8)
cv2_imshow(mask)
cv2.rectangle(mask,(100, 100), (400, 300), 255, -1)
bitwise_and = cv2.bitwise_and(image, image, mask=mask)
cv2_imshow(bitwise_and)
bitwise_or = cv2.bitwise_or(image, image, mask=mask)
cv2_imshow( bitwise_or)
bitwise_xor = cv2.bitwise_xor(image, image, mask=mask)
cv2_imshow( bitwise_xor)
bitwise_not = cv2.bitwise_not(image, mask=mask)
cv2_imshow( bitwise_not)




cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure(figsize=(30, 15))

rows = 2

col = 6

fig.add_subplot(rows, col,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("First")

fig.add_subplot(rows, col,2)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Second")

fig.add_subplot(rows, col,3)
plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Third")

fig.add_subplot(rows, col,4)
plt.imshow(cv2.cvtColor(bitwise_or, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Fourth")

fig.add_subplot(rows, col,5)
plt.imshow(cv2.cvtColor(bitwise_xor, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Fifth")

fig.add_subplot(rows, col,6)
plt.imshow(cv2.cvtColor(bitwise_not, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Sixth")
```
