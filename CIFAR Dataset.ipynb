# TASK 1

```python
import os

dataset_dir = r'Pets_Emotions'

class_names = os.listdir(dataset_dir)
class_counts = {}

for class_name in class_names:
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        class_count = len(os.listdir(class_path))
        class_counts[class_name] = class_count


for class_name, count in class_counts.items():
    print(f"Class '{class_name}': {count} samples")
```

# TASK 2

```python
import cv2
import numpy as np
from sklearn.model_selection import train_test_split


target_size = (224, 224)

data = []
labels = []

class_names = os.listdir(dataset_dir)
class_mapping = {class_name: idx for idx, class_name in enumerate(class_names)}

for class_name in class_names:
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        for img_file in os.listdir(class_path):
            img_path = os.path.join(class_path, img_file)
            img = cv2.imread(img_path)
            img = cv2.resize(img, target_size)
            img = img/255
            data.append(img)
            labels.append(class_mapping[class_name])

data = np.array(data)
labels = np.array(labels)

X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)


print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# TASK 3

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Class': [], 'Count': []}
class_names = os.listdir(dataset_dir)

for class_name in class_names:
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        num_samples = len(os.listdir(class_path))
        data['Class'].append(class_name)
        data['Count'].append(num_samples)

df = pd.DataFrame(data)

df = df.sort_values(by='Count', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(df['Class'], df['Count'])
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Visualization of Pet Emotions Dataset')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()

plt.show()
```

# TASK 4

```python
import random
num_samples_per_class = 5

class_mapping = {class_name: idx for idx, class_name in enumerate(os.listdir(dataset_dir))}

fig, axes = plt.subplots(len(class_mapping), num_samples_per_class, figsize=(12, 8))

for class_name, class_idx in class_mapping.items():
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        images = os.listdir(class_path)
        random.shuffle(images)
        selected_samples = images[:num_samples_per_class]

        for i, sample_file in enumerate(selected_samples):
            img_path = os.path.join(class_path, sample_file)
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            axes[class_idx, i].imshow(img)
            axes[class_idx, i].set_title(class_name)
            axes[class_idx, i].axis('off')

plt.tight_layout()
plt.show()
```

# TASK 5

```python
class_mapping = {class_name: idx for idx, class_name in enumerate(os.listdir(dataset_dir))}
class_data = {class_name: [] for class_name in class_mapping}
for class_name, class_idx in class_mapping.items():
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        for image_file in os.listdir(class_path):
            image_path = os.path.join(class_path, image_file)
            img = cv2.imread(image_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            class_data[class_name].append(img)

class_statistics = {}

for class_name, data in class_data.items():
    num_samples = len(data)
    mean_image = np.mean(data, axis=0)
    std_image = np.std(data, axis=0)
    min_image = np.min(data, axis=0)
    max_image = np.max(data, axis=0)

    class_statistics[class_name] = {
        "Num Samples": num_samples,
        "Mean Image": mean_image,
        "Std Deviation Image": std_image,
        "Min Image": min_image,
        "Max Image": max_image,
    }

for class_name, stats in class_statistics.items():
    print("===================================================")
    print(f"Class: {class_name}")
    print(f"Number of Samples: {stats['Num Samples']}")
    print(f"Mean Image: {np.mean(stats['Mean Image'])}")
    print(f"Standard Deviation Image: {np.mean(stats['Std Deviation Image'])}")
    print(f"Min Image: {np.mean(stats['Min Image'])}")
    print(f"Max Image: {np.mean(stats['Max Image'])}")
```
